```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%load_ext autoreload
%autoreload 2
%matplotlib notebook 
```

```python
# Global Parameters
data_path = './dataset_1.csv'
n_epochs = 1000
learning_rate = 0.001
random = False
np.random.seed(1000)
```

```python
class Perceptron:
    def __init__(self):
        # Initialize the parameters
        self.weights = None
        self.weights_list = None
    
    def extend_vector(self, mat):
        new_mat = np.ones((len(mat), len(mat[0])+1))
        new_mat[:, 1:] = mat
        return new_mat
    
    def activation(self, input_x):
        if np.dot(self.weights, input_x) >= 0:
            return 1
        else:
            return 0
    
    def metrics(self, predicted, actual):
        confusion_matrix = np.zeros((2,2))
        N = len(actual)
        for i in range(N):
            confusion_matrix[int(actual[i]),int(predicted[i])] += 1
        
        accuracy = (confusion_matrix[0, 0] + confusion_matrix[1, 1])/N
        precision = (confusion_matrix[1, 1])/(confusion_matrix[1, 1] + confusion_matrix[0, 1])
        recall = (confusion_matrix[1, 1])/(confusion_matrix[1, 1] + confusion_matrix[1, 0])
        f1 = 2*(precision*recall)/(precision+recall)
        
        return np.array([accuracy, precision, recall, f1])
    
    def predict(self, input_features):
        extend_vector = self.extend_vector(input_features)
        predictions = []
        for j in range(len(extend_vector)):
            predictions.append(self.activation(extend_vector[j]))
        
        return np.array(predictions)
         
    def fit(self, input_features, labels, n_epochs= 1, learning_rate= 1, random=False, verbose=10):
        # Append column of ones 
        extended_input = self.extend_vector(input_features)
        N = len(extended_input)
        
        # History of weights
        weights_list = []
        
        # Initialization of weights
        if random:
            self.weights = np.random.rand((3))
        else:
            self.weights = np.zeros(3)
        
        # Maximum Accuracy obtained
        max_accuracy = 0
        
        for i in range(n_epochs):
            for j in range(N):
                output = self.activation(extended_input[j])
                self.weights = self.weights + (labels[j]-output)*learning_rate*extended_input[j]
            
            # Store the weights
            weights_list.append(self.weights)
            
            # Predict on the dataset
            predicted = self.predict(input_features)
            
            # Get performance scores
            temp_metrics = self.metrics(predicted, labels)
            accuracy = temp_metrics[0]
            
            if accuracy > max_accuracy:
                max_accuracy = accuracy
                best_weights = self.weights
                if max_accuracy == 1:
                    break
            
            if verbose:
                if i%verbose == 0:
                    print('Epoch '+str(i+1)+':')
                    print('-------')
                    print('Accuracy: '+str(temp_metrics[0]))
                    print('Precision: '+ str(temp_metrics[1]))
                    print('Recall: '+str(temp_metrics[2]))
                    print('F1: '+str(temp_metrics[3]))
                    print('\n')
        
        # Update model weights to the best weights obtained
        self.weights = best_weights
        self.weights_list = weights_list
        
        predicted = self.predict(input_features)
        temp_metrics = self.metrics(predicted, labels)
        print('Epoch '+str(i+1)+':')
        print('-------')
        print('Accuracy: '+str(temp_metrics[0]))
        print('Precision: '+ str(temp_metrics[1]))
        print('Recall: '+str(temp_metrics[2]))
        print('F1: '+str(temp_metrics[3]))
        print('\n')
```

```python
def train_test_split(data, split=0.8):
    np.random.shuffle(data)
    N = len(data)
    train = data[0:int(split*N)]
    test  = data[int(split*N):N]
    
    X_train = train[:, :2]
    Y_train = train[:, 2]
    
    X_test = test[:, :2]
    Y_test = test[:, 2]
    
    return X_train, X_test, Y_train, Y_test, train, test
```

```python
def visualize(data_array, labels, axis, weights=None):
    axis.scatter(data_array[:, 0], data_array[:, 1], c=labels, alpha=0.7)
    
    if weights is not None:
        X1 = data_array[:, 0].min()
        X2 = data_array[:, 0].max()
        Y1 = -(weights[0]+weights[1]*X1)/weights[2]
        Y2 = -(weights[0]+weights[1]*X2)/weights[2]
        
        axis.plot([X1, X2], [Y1, Y2])
```

```python
# Load Dataset
data = pd.read_csv(data_path, names=['id', 'f1', 'f2', 'label'])
data = data.drop('id', axis=1)
data_array = data.values
```

```python
visualize(data_array[:, 0:2], data_array[:, 2])
```

```python
X_train, X_test, Y_train, Y_test, train, test = train_test_split(data_array)
```

```python
# Visualize Training Data
# plt.scatter(X_train[:, 0], X_train[:, 1], c=Y_train, alpha=0.7)
plt.scatter(X_test[:, 0], X_test[:, 1], c=Y_test, alpha=0.7)
```

```python
# Train Model
model = Perceptron()
model.fit(X_train, Y_train, n_epochs=n_epochs, learning_rate=learning_rate, random=random, verbose=2)
weights_list = model.weights_list
```

```python

fig = plt.figure()
ax = fig.add_subplot(122)
plt.ion()

fig.show()
fig.canvas.draw()

i=1
for weights in weights_list:
    ax.clear()
    
    predict = model.predict(X_test)
    ax.scatter(X_test[:, 0], X_test[:, 1], c=predict, alpha=0.7)
    
    X1 = data_array[:, 0].min()
    X2 = data_array[:, 0].max()
    Y1 = -(weights[0]+weights[1]*X1)/weights[2]
    Y2 = -(weights[0]+weights[1]*X2)/weights[2]  
    ax.plot([X1, X2], [Y1, Y2])
    ax.set_title('Epoch: '+str(i))
    i+=1 
    
    plt.pause(0.05)
    fig.canvas.draw()
```
